```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la funci√≥n de costo y su derivada
def costo(x):
    return x ** 2 - 4 * x + 6


def derivada_costo(x):
    return 2 * x - 4


def segunda_derivada_costo(x):
    return 2  # Es una constante positiva
```

```python
# Crear un rango de valores para x
x = np.linspace(0, 4, 400)  # Ajusta el rango seg√∫n sea necesario

# Calcular los valores de y para cada funci√≥n
y_costo = costo(x)
y_derivada_costo = derivada_costo(x)

# Crear las gr√°ficas
plt.figure(figsize=(12, 6))

# Graficar la funci√≥n de costo
plt.subplot(2, 1, 1)
plt.plot(x, y_costo, label='Costo: $x^2 - 4x + 6$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([2], [2], color='red', zorder=5)  # Mostrar el m√≠nimo en rojo
plt.title('Funci√≥n de Costo')
plt.xlabel('Cantidad de Producci√≥n (x)')
plt.ylabel('Costo (C(x))')
plt.legend()

# Graficar la derivada de la funci√≥n de costo
plt.subplot(2, 1, 2)
plt.plot(x, y_derivada_costo, label='Derivada del Costo: $2x - 4$', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Derivada de la Funci√≥n de Costo')
plt.xlabel('Cantidad de Producci√≥n (x)')
plt.ylabel('Derivada (C\'(x))')
plt.legend()

# Ajustar la visualizaci√≥n
plt.tight_layout()
plt.show()
```
