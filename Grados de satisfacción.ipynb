# Ejemplo 1
Tenemos una muestra de 25 datos de una variable aleatoria que es el grado de satisfacción de los clientes de un hotel

```python
import matplotlib.pyplot as plt
```

```python
grados_satisfaccion = ["Muy satisfecho", "Satisfecho", "Medianamente satisfecho", "Descontento"]
frecuencias = [10, 7, 5, 3]
```

### Primero se representará con un gráfico de tarta o pastel

```python
plt.pie(frecuencias, labels = grados_satisfaccion, autopct = "%1.1f%%", shadow = True, 
        explode = (0.1,0,0, 0))
plt.title("Grado de satisfacción de los clientes de un hotel")
plt.show()
```

### A continuación se representa la información con un gráfico de barras

```python
plt.figure(figsize=(10,6))
plt.bar(grados_satisfaccion, frecuencias, color="orchid")
plt.title("Grado de satisfacción de los clientes de un hotel")
plt.ylabel("Número de clientes satisfechos")
plt.show()
```
