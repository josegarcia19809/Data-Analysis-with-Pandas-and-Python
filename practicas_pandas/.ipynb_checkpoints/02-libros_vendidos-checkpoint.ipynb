## Selección de columnas
### Usando la notación punto

```python
libros.Titulo
```

```python
# Pero no siempre funcionará por los espacios en el nombre de la columna
libros.Calificación de usuarios
```

```python
# En su lugar se ocupará acceso a columna por nombre
libros["Calificación de usuarios"]
```

```python
# Se puede usar una variable para guardar los datos de una columna
calificaciones = libros["Calificación de usuarios"]
calificaciones
```

### Selección de varias columnas

```python
titulo_y_anio_lanzamiento = libros[["Titulo", "Año"]]
titulo_y_anio_lanzamiento
```

## Selección de renglones

```python
# Seleccionar el renglón que tiene el índice 5
libros.iloc[5]
```

```python
# También se puede guardar la selección de renglones en variables
mi_libro_favorito = libros.iloc[0]
mi_libro_favorito
```

### Seleccionar múltiples renglones

```python
# Seleccionar los renglones que van desde el índice 3 hasta el 7
varios_libros = libros.iloc[3:8]
varios_libros
```

```python
# Seleccionar los renglones que van desde el índice 0 hasta el 4
varios_libros = libros.iloc[:5]
varios_libros
```

```python
# Seleccionar los últimos 3 renglones
varios_libros = libros.iloc[-3:]
varios_libros
```

```python
# Seleccionar los renglones que van desde el índice 0 hasta el 9, de 3 en 3
varios_libros = libros.iloc[0:10:3]
varios_libros
```
