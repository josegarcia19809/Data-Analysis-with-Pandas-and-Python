En este curso usaremos tres paquetes desarrollados por la comunidad científica:

    numpy: es una biblioteca que proporciona soporte para arrays multidimensionales y funciones matemáticas para operaciones numéricas eficientes. Es la base de muchas otras bibliotecas de análisis de datos en Python. NumPy es especialmente útil para realizar cálculos numéricos y operaciones de álgebra lineal. Ofrece una amplia gama de funciones para manejar datos numéricos y realizar cálculos numéricos avanzados.
    
    matplotlib: es una biblioteca de visualización en 2D que permite crear una variedad de gráficos y visualizaciones. Proporciona una interfaz similar a la de MATLAB y es útil para crear gráficos de líneas, histogramas, gráficos de dispersión, gráficos de barras, entre otros. Matplotlib te permite personalizar y configurar gráficos según tus necesidades. Además, existen interfaces de alto nivel como Seaborn que se basan en Matplotlib y facilitan la creación de gráficos más atractivos y complejos.

    pandas: es una biblioteca de manipulación y análisis de datos que proporciona estructuras de datos flexibles y herramientas para trabajar con conjuntos de datos tabulares. La estructura principal en Pandas es el DataFrame, que te permite almacenar y manipular datos de manera eficiente. Pandas es ampliamente utilizado para limpiar, transformar y explorar datos antes de realizar análisis estadísticos. También es compatible con la carga y escritura de datos desde y hacia diferentes formatos, como CSV, Excel y bases de datos.

```python
import numpy as np # importa numpy
import matplotlib.pyplot as plt #importa las herramientas para graficar
import pandas as pd # importa pandas
```

A su vez, podemos crear en nuestro directorio raíz un archivo CSV (data.csv) con datos de ejemplo. Tomar como referencia lo siguiente:
Nota. Recuerde que para crear el archivo debe dar clic en el menú "File", "New", "Text File" donde deberá poner la información y guardar como csv

Una vez teniendo nuestro archivo, cargaremos nuestro conjunto de datos en un DataFrame de pandas:

```python
data = pd.read_csv('data.csv')

print(data)
```

 Al dar clic en "run" veremos la tabla en pantalla.
 
 Procedemos a calcular la media, mediana y desviación estándar con numpy:

```python
mean_age = np.mean(data['age'])
median_age = np.median(data['age'])
des_age = np.std(data['age'])

print("La media es: ", mean_age) # 25.2
print("La mediana es: ", median_age) # 25.0
print("La desviación estándar es: ", des_age) # 2.1354156504062622
```

Por último, crearemos un gráfico de barras simple con matplotlib para visualizar nuestros datos:

```python
plt.bar(data['name'], data['age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age of persons')
plt.xticks(rotation=45)
plt.show()
```

¿Qué pasa si agregamos 'h' en la primera instrucción?
Nota. Podemos cambiar el color de las barras agregando la instrucción: color 'darkseagreen' (en la primera línea).


```python
plt.barh(data['name'], data['age'],color='darkseagreen')
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age of persons')
plt.xticks(rotation=45)
plt.show()
```
