```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Crear un rango de valores para x
x = np.linspace(0, 6, 13)  # Ajusta el rango seg√∫n sea necesario

# Calcular los valores de y
y = -2 * x ** 2 + 12 * x - 15
```

```python
# Imprimir valores
for xi, yi in zip(x, y):
    print(f"x: {xi:.2f}, f(x): {yi:.2f}")
```

```python
# Crear las gr√°ficas
plt.figure(figsize=(12, 6))

# Graficar la funci√≥n de ganancia
plt.plot(x, y, label='Ganancia: $-2x^2 + 12x - 15$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([3], [3], color='red', zorder=5)  # Mostrar el m√°ximo en rojo
plt.title('Funci√≥n de Ganancia')
plt.xlabel('N√∫mero de Unidades Vendidas (x)')
plt.ylabel('Ganancia (G(x))')
plt.legend()

# Ajustar la visualizaci√≥n
plt.tight_layout()
plt.show()
```
