```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la función de costo y su derivada
def costo(x):
    return x ** 2 - 4 * x + 6


def derivada_costo(x):
    return 2 * x - 4


def segunda_derivada_costo(x):
    return 2  # Es una constante positiva
```

```python
# Crear un rango de valores para x
x = np.linspace(0, 4, 400)  # Ajusta el rango según sea necesario

# Calcular los valores de y para cada función
y_costo = costo(x)
y_derivada_costo = derivada_costo(x)

# Crear las gráficas
plt.figure(figsize=(12, 6))

# Graficar la función de costo
plt.subplot(2, 1, 1)
plt.plot(x, y_costo, label='Costo: $x^2 - 4x + 6$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([2], [2], color='red', zorder=5)  # Mostrar el mínimo en rojo
plt.title('Función de Costo')
plt.xlabel('Cantidad de Producción (x)')
plt.ylabel('Costo (C(x))')
plt.legend()

# Graficar la derivada de la función de costo
plt.subplot(2, 1, 2)
plt.plot(x, y_derivada_costo, label='Derivada del Costo: $2x - 4$', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Derivada de la Función de Costo')
plt.xlabel('Cantidad de Producción (x)')
plt.ylabel('Derivada (C\'(x))')
plt.legend()

# Ajustar la visualización
plt.tight_layout()
plt.show()
```
