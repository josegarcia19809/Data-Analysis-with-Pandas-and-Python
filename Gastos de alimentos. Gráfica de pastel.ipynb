# Gastos de alimentos
Se quiere saber cu√°nto gasta una familia en diferentes alimentos

```python
import matplotlib.pyplot as plt
```

```python
alimentos = ["Pavo", "Papas", "Pastel de calabaza", "Relleno"]
gastos = [26.0, 3.50, 9.50, 7.00]
```

```python
plt.pie(gastos, labels = alimentos, autopct = "%1.1f%%", shadow = True, 
        explode = (0,0,0.1, 0))
plt.title("Gastos en alimentos")
plt.show()
```
