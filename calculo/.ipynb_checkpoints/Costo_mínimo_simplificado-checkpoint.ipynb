```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Crear un rango de valores para x
x = np.linspace(0, 4, 9)  # Ajusta el rango según sea necesario

# Calcular los valores de y
y = x ** 2 - 4 * x + 6

# Crear las gráficas
plt.figure(figsize=(12, 6))

# Graficar la función de costo
plt.plot(x, y, label='Costo: $x^2 - 4x + 6$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.scatter([2], [2], color='red', zorder=5)  # Mostrar el mínimo en rojo
plt.title('Función de Costo')
plt.xlabel('Cantidad de Producción (x)')
plt.ylabel('Costo (C(x))')
plt.legend()

# Ajustar la visualización
plt.tight_layout()
plt.show()
```

```python
# Imprimir valores
for xi, yi in zip(x, y):
    print(f"x: {xi:.2f}, f(x): {yi:.2f}")
```
