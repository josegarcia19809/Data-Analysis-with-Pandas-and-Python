### Agregando valores y etiquetas a los ejes

```python
plt.plot([1,2,3,4])
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, "ro")
plt.show()
```

```python
plt.plot(x, y)
plt.axis([0, 6, 0, 20])
plt.show()
```
