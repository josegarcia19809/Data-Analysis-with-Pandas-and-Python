```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definir el rango de x
x = np.arange(0, 501, 50)  # De 0 a 500 en intervalos de 50

# Definir la función f(x)
def f(x):
    return 1.15 * x

# Calcular f(x)
y = f(x)
```

```python
# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Gráfica de la función f(x) = 1.15x')
plt.xlabel('Cuenta ($)')
plt.ylabel('Total a Pagar ($)')
plt.grid(True)
plt.xticks(np.arange(0, 501, 50))
plt.yticks(np.arange(0, 575, 50))
plt.show()
```
