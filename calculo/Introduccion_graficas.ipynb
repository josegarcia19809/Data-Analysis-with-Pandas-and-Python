### Agregando valores y etiquetas a los ejes

```python
plt.plot([1, 2, 3, 4])
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, "ro")
plt.show()
```

```python
plt.plot(x, y)
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import numpy as np
```

```python
data = np.arange(0.0, 10.0, 0.5)
data
```

```python
plt.plot(data, data ** 2, "g--")
plt.show()
```

```python
plt.plot(data, data ** 2, linewidth=3)
plt.show()
```

```python
line, = plt.plot(data, data ** 2, linewidth=3)
line.set_antialiased(False)
plt.show()
```

```python
lines = plt.plot(data, data, data, data ** 2)
plt.setp(lines, color="r", linewidth=2.0)
plt.show()
```

```python
plt.plot(data, data ** 2, alpha=.3)
plt.show()
```

```python
plt.plot(data, data ** 2, marker="+", linestyle="-.")
plt.show()
```

```python
plt.plot(data, data ** 2, marker="+", linestyle=":")
plt.show()
```

```python
x = np.arange(0.0, 4.5, 0.5)
```
