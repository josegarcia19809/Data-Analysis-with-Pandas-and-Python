# Vivero

Se cultivan Espinaca, Nabo, Ruibarbo, Brócoli,
Col rizada. Se quedan como plántulas.
Algunas plántulas sobreviven y algunas no.
Se tiene un registro de ello, y se quiere graficar
dichos datos.

```python
import matplotlib.pyplot as plt
```

```python
plantas = ["Espinaca", "Nabo", "Ruibarbo", "Brócoli", "Col rizada"]
murieron = [10, 25, 5, 30, 21]
germinaron = [74, 88, 56, 69, 59]
```

### Se crea un gráfico de barras con las plantas que sobrevivieron

```python
plt.figure(figsize=(6,6))
plt.bar(plantas, germinaron, color="orchid")
plt.title("Tasas de germinación")
plt.ylabel("Número de plántulas")
plt.show()
```

### Se crea un gráfico de barras con las plantas que murieron

```python
plt.figure(figsize=(6,6))
plt.bar(plantas, murieron, color="red")
plt.title("Tasas de fallo")
plt.ylabel("Número de plántulas que murieron")
plt.show()
```

### En un solo gráfico se pueden combinar ambos tipos de plantas
### Las barras van lado a lado

```python
plt.figure(figsize=(6,6))
plt.title("Plántulas de invernadero")
plt.ylabel("Número de plántulas")
plt.bar(plantas, murieron, color="#3498db", label="Plántulas que murieron")
plt.bar(plantas, germinaron, width=0.4, color="#2ecc71", align="edge" , label="Plántulas que sobrevivieron")
plt.legend(shadow=True, frameon=True, facecolor="white")
plt.show()
```

### Existe un formato horizontal

```python
plt.figure(figsize=(6,6))
plt.title("Plantas que sobrevivieron")
plt.xlabel("Número de plantas")
plt.barh(plantas, germinaron, color = "#2ecc71")
plt.show()
```
