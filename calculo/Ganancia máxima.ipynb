```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir la funci√≥n de ganancia y su derivada
def ganancia(x):
    return -2 * x**2 + 12 * x - 15

def derivada(x):
    return -4 * x + 12
```

```python
# Crear un rango de valores para x
x = np.linspace(0, 6, 400)  # Ajusta el rango seg√∫n sea necesario

# Calcular los valores de y para cada funci√≥n
y_ganancia = ganancia(x)
y_derivada = derivada(x)

# Crear las gr√°ficas
plt.figure(figsize=(10, 6))

# Graficar la funci√≥n de ganancia
plt.subplot(2, 1, 1)
plt.plot(x, y_ganancia, label='Ganancia: $-2x^2 + 12x - 15$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Funci√≥n de Ganancia')
plt.xlabel('N√∫mero de Unidades Vendidas (x)')
plt.ylabel('Ganancia (G(x))')
plt.legend()

# Graficar la derivada de la funci√≥n de ganancia
plt.subplot(2, 1, 2)
plt.plot(x, y_derivada, label='Derivada: $-4x + 12$', color='red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Derivada de la Ganancia')
plt.xlabel('N√∫mero de Unidades Vendidas (x)')
plt.ylabel('Derivada (G\'(x))')
plt.legend()

# Ajustar la visualizaci√≥n
plt.tight_layout()
plt.show()
```

```python
# Definir la segunda derivada como una constante
def segunda_derivada(x):
    return -4 * np.ones_like(x)  # Crear un array del mismo tama√±o que x con valor -4

# Crear un rango de valores para x
x = np.linspace(0, 6, 400)  # Ajusta el rango seg√∫n sea necesario

# Calcular los valores de y para la segunda derivada
y_segunda_derivada = segunda_derivada(x)

# Crear la gr√°fica
plt.figure(figsize=(8, 4))

# Graficar la segunda derivada
plt.plot(x, y_segunda_derivada, label='Segunda Derivada: $-4$', color='green')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Segunda Derivada de la Ganancia')
plt.xlabel('N√∫mero de Unidades Vendidas (x)')
plt.ylabel('Segunda Derivada (G\'\'(x))')
plt.legend()

# Ajustar la visualizaci√≥n
plt.tight_layout()
plt.show()
```
