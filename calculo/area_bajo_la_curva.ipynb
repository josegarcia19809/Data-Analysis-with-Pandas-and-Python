```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
# Definir la función que queremos integrar
def f(x):
    return x**2 + 1

# Definir los límites de integración
a = 0  # límite inferior
b = 3  # límite superior
```

```python
# Calcular el área bajo la curva usando scipy.integrate.quad
area, error = quad(f, a, b)

# Imprimir el resultado
print("El área bajo la curva es: ", area, "unidades cuadradas")
```

```python
# Crear un array de valores de x
x = np.linspace(0, 3, 400)

# Evaluar la función en los valores de x
y = f(x)

# Graficar la función
plt.plot(x, y, label=r'$y = x^2 + 1$', color='blue')

# Rellenar el área bajo la curva entre x=0 y x=3
plt.fill_between(x, y, where=[(xi >= 0) and (xi <= 3) for xi in x], color='skyblue', alpha=0.4)

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Área bajo la curva $y = x^2 + 1$')
plt.legend()

# Mostrar la gráfica
plt.show()
```
