# Inventario Jardiner√≠a Edo Mex

```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

```python
inventario = pd.read_csv("datax/inventario_edo_mex.csv")
inventario
```

```python
# Obtenemos la frecuencia de la columna 'tipo_de_producto'
frecuencia_tipo_de_producto = inventario['tipo de producto'].value_counts()

print(frecuencia_tipo_de_producto)
```

```python
frecuencia_tipo_de_producto.plot(kind='bar', color='skyblue')
plt.title('Frecuencia de tipo de producto')
plt.xlabel('Tipo de producto')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Graficamos un gr√°fico de pastel mostrando los porcentajes
plt.pie(frecuencia_tipo_de_producto, labels=frecuencia_tipo_de_producto.index, 
        autopct='%1.1f%%', colors=['red', 'blue', 'green', "yellow"])
plt.title('Distribuci√≥n de tipos de producto')
plt.show()
```

```python
precios = inventario["precio"]
precios.sort_values()
```

```python
precios = inventario["precio"]
# Media
media_precios = np.mean(precios)
# Mediana
mediana_precios = np.median(precios)
print(f"media_precios: {media_precios}")
print(f"mediana_precios: {mediana_precios}")
```

```python
cantidades = inventario["cantidad"]
# Media
media_cantidades = np.mean(cantidades)
# Calculamos la desviaci√≥n est√°ndar
desviacion_estandar_cantidades = statistics.stdev(cantidades)
print(f"media_cantidades: {media_cantidades}")
print(f"desviacion_estandar_cantidades: {desviacion_estandar_cantidades:.2f}")
```

```python
# Definir los l√≠mites de clase
bins = [20, 30, 40, 50, 60]
# Usamos pd.cut() para segmentar los datos en los intervalos
precios_binned = pd.cut(precios, bins)

# Creamos la tabla de distribuci√≥n de frecuencias usando value_counts()
tabla_frecuencia = precios_binned.value_counts()

print(tabla_frecuencia.sort_index())
```

```python
# Crear el histograma
# Graficamos el histograma
tabla_frecuencia.sort_index().plot(kind='bar', color='lightblue', edgecolor='black')


# A√±adir etiquetas y t√≠tulo
plt.title('Histograma de Frecuencias')
plt.xlabel('Clases')
plt.ylabel('Frecuencia')

# Mostrar el gr√°fico
plt.show()
```
