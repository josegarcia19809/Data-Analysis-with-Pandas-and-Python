```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definir el rango de valores para x, evitando el punto de indeterminación
x = np.linspace(0, 4, 400)
x = x[x != 2]  # Evitar el punto x = 2

# Definir la función
def f(x):
    return 2 * (x + 2)

# Calcular los valores de la función
y = f(x)
```

```python
# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = 2(x + 2)$', color='blue')
plt.axvline(x=2, color='red', linestyle='--', label='x = 2 (indeterminación)')
plt.scatter([2], [8], color='red', zorder=5, label='Límite en $x = 2$')

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la función $f(x) = 2(x + 2)$')
plt.legend()
plt.grid(True)
plt.yticks(np.arange(0, 20, 1))
plt.ylim(0, 20)  # Ajustar el rango del eje y para mejor visualización

# Mostrar la gráfica
plt.show()
```
